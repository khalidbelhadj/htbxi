# Converting Borrows

```python
import geopandas as gpd
import os

# Input and output folders
input_folder = "msoa2021"
output_folder = "out"

# Create output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Process all shapefiles in the folder
for filename in os.listdir(input_folder):
    if filename.endswith(".shp"):
        # Read shapefile
        shapefile_path = os.path.join(input_folder, filename)
        gdf = gpd.read_file(shapefile_path)
        
        # Reproject to WGS84 (long/lat)
        gdf_wgs84 = gdf.to_crs(epsg=4326)
        
        # Save to new location
        output_path = os.path.join(output_folder, filename)
        gdf_wgs84.to_file(output_path)
        
        print(f"Converted {filename}")
```

```python
import matplotlib.pyplot as plt

# Read the converted shapefile
converted_file = "out/Barnet.shp"
gdf = gpd.read_file(converted_file)

# Check the coordinate reference system
print("CRS:", gdf.crs)  # Should show EPSG:4326 or similar WGS84 reference

# Check some coordinates to verify they're in longitude/latitude range
print("Sample coordinates:")
print(gdf.geometry.iloc[0].representative_point().coords[0])  # Should be in decimal degrees

# Plot to visually inspect
fig, ax = plt.subplots(figsize=(10, 6))
gdf.plot(ax=ax)
plt.title("Converted Shapefile (Should look normal in WGS84)")
plt.show()
```
